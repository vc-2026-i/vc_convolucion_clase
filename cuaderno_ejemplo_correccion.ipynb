```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('image1.jpg')
img2 = cv2.imread('image2.jpg')
```

```python
hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 5))
plt.plot(hist1, color='blue')
plt.title('Histograma de la imagen 1')
```

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img1_binary = cv2.threshold(img1_gray, 127, 255, cv2.THRESH_BINARY)[1]
img2_binary = cv2.threshold(img2_gray, 127, 255, cv2.THRESH_BINARY)[1]
```

```python
hist1_bin = cv2.calcHist([img1_binary], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 5))
plt.plot(hist1_bin, color='blue')
plt.title('Histograma de la imagen 1 binaria')
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1_binary, cmap='gray')
plt.title('Imagen 1 binaria')
plt.subplot(1, 2, 2)
plt.imshow(img2_binary, cmap='gray')
plt.title('Imagen 2 binaria')
```

```python
img1_erosion = cv2.erode(img1_binary, None, iterations=1)
plt.figure(figsize=(10, 5))
plt.imshow(img1_erosion, cmap='gray')
plt.title('Imagen 1 después de erosión')
```

```python
img1_opens = cv2.morphologyEx(img1_binary, cv2.MORPH_OPEN, None)
plt.figure(figsize=(10, 5))
plt.imshow(img1_opens, cmap='gray')
plt.title('Imagen 1 después de apertura')
```

```python
img2_erosion = cv2.erode(img2_binary, None, iterations=1)
plt.figure(figsize=(10, 5))
plt.imshow(img2_erosion, cmap='gray')
plt.title('Imagen 2 después de erosión')
```

```python
img2_dilatacion = cv2.dilate(img2_erosion, None, iterations=2)
plt.figure(figsize=(10, 5))
plt.imshow(img2_dilatacion, cmap='gray')
plt.title('Imagen 2 después de una erosion y dos dilataciones')
```

```python
img3 = cv2.imread('image3.jpg')
img3_gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))
plt.imshow(img3_gray, cmap='gray')
plt.title('Imagen 3 en escala de grises')
```

```python
img3_binary = cv2.threshold(img3_gray, 165, 200, cv2.THRESH_BINARY)[1]
plt.figure(figsize=(10, 5))
plt.imshow(img3_binary, cmap='gray')
plt.title('Imagen 3 binaria')
```

```python
img3_erosion = cv2.erode(img3_binary, None, iterations=2)
plt.figure(figsize=(10, 5))
plt.imshow(img3_erosion, cmap='gray')
plt.title('Imagen 3 después de erosión')
```

```python
img3_dilatacion = cv2.dilate(img3_erosion, None, iterations=38)
plt.figure(figsize=(10, 5))
plt.imshow(img3_dilatacion, cmap='gray')
plt.title('Imagen 3 después de erosiones y una dilataciones')
```

```python
# Aplicamos componentes conexas a la imagen 3 después de las erosiones y dilataciones
num_labels, labels_im = cv2.connectedComponents(img3_dilatacion)
```

```python
# Visualizamos las componentes conexas
plt.figure(figsize=(10, 5))
plt.imshow(labels_im, cmap='nipy_spectral')
plt.title(f'Componentes conexas en la imagen 3 (Número de componentes: {num_labels})')
```
